```python
#importing the pandas
import pandas as pd
df = pd.read_excel(r'C:\Users\ragha\Downloads\Online Retail.xlsx')
df
```

```python
#converting the datetime64 datatype to int
df['day'] = df['InvoiceDate'].dt.day
df['month'] = df['InvoiceDate'].dt.month
df['year'] = df['InvoiceDate'].dt.year
df
```

```python
#dropping the null value from customer ID
df.dropna(subset = ['CustomerID'],inplace = True)
df['CustomerID'].isnull().sum()
```

```python
#EDA
df.info()
```

```python
#the necessary changes
df['CustomerID'] = df['CustomerID'].astype(int)
df['day'] = df['day'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)


df.info()
```

```python
#importing the linear model to fit or train the dataset
from sklearn import linear_model
from sklearn.linear_model import LinearRegression


#fitted the dataset
quantity_prediction = linear_model.LinearRegression()
quantity_prediction.fit(df[['CustomerID','day','month','year']],df.Quantity)
```

```python
#ignoring the warnings
import warnings
warnings.filterwarnings('ignore')

quantity_prediction.predict([[12680,9,12,2011]])
```
